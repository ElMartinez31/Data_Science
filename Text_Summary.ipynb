```python
#Now let's vectorize

tokenize = TfidfVectorizer( #vectorizer is a better name than tokenize
    stop_words = stopwords.words('spanish'),
    norm = 'l1'
    )
```

```python

X = tokenize.fit_transform(lines)
X
X.shape # no ()
#95 lines (tokens), 720 different words

#X est une matrice sparse (éparse), où chaque ligne correspond à un document
# ou une phrase, et chaque colonne à un mot unique (ou un token unique).
```

```python
rows, columns = X.shape

for i in range(rows):
  for j in range(columns):
    print(X[i,j], end=' ')
  print()
#Here we print the value of each word in each document,
#that's why it's a sparse matrix
```

```python
# Now let's score tfidf each row

def get_score(tfidf_row):
    x = tfidf_row[tfidf_row != 0]
    return x.mean()


def get_score(tfidf_row): #Gemini version after the division by 0 proc error
    x = tfidf_row[tfidf_row != 0]
    # Check if 'x' is empty before calculating the mean
    if x.size == 0:
        return 0  # Return 0 if 'x' is empty to avoid division by zero
    else:
        return x.mean()
```

```python
# Now let's use that on every row of lines
# First, instantiate a null series of the size of lines (95)
# in order to store the scores
scores = np.zeros(len(lines))
scores
```

```python
scores = np.zeros(len(lines))
for i in range(len(lines)):
    score = get_score(X[i,:])#score the whole line of X
    scores[i] = score
```

```python
scores #95 values, each corresponding to the mean tfidf score of each row of lines
```

```python
sort_idx = np.argsort(-scores)# get the indices of the sorted (deacreasing) values of scores
sort_idx
```

```python
print('Summary:')
sentences = []
for i in range(0,7):
  sentences.append([sort_idx[i], scores[sort_idx[i]], lines[sort_idx[i]]])
  print(f"{scores[sort_idx[i]]} : {lines[sort_idx[i]]}")
```

```python
sentences
```

```python
# This 'summary' looks a little bit messy. The idea now
# would be to take the very first indice of these tokens (the indice of lines of this sentence)
# and sort it with this new index (we used the first inedexes to get the most important tokens)
# now we have them, we sort them in the original way

sentences_sorted = sorted(sentences, key = lambda x: x[0])
sentences_sorted
```
