1. FastAPI
À quoi ça sert ?

Framework moderne pour construire des APIs RESTful en Python.

Fonctionnalités clés :

Validation automatique des données grâce aux type hints

Génération automatique de documentation interactive

Performances élevées (basé sur Starlette et Pydantic)

Support natif pour les opérations asynchrones


2. Uvicorn

À quoi ça sert ?

Serveur ASGI (Asynchronous Server Gateway Interface) ultra-rapide.

Pourquoi l'utiliser ?

Spécialement optimisé pour FastAPI

Supporte le code asynchrone


3. Ngrok

À quoi ça sert ?

Outil qui crée un tunnel sécurisé entre internet et votre machine locale.

Utilité dans votre projet :

Permet d'exposer votre API locale (sur Colab) sur internet

Génère une URL publique (ex: https://xxxx.ngrok.io)

Évite de configurer un vrai serveur cloud

Léger et performant


4. Asyncio

À quoi ça sert ?

Bibliothèque pour la programmation asynchrone en Python.

Rôle dans FastAPI :

Permet de gérer plusieurs requêtes simultanément

Améliore les performances pour les opérations I/O (BDD, API externes)




5. CORS (Cross-Origin Resource Sharing)

À quoi ça sert ?

Mécanisme de sécurité qui contrôle les requêtes entre différents domaines.

Pourquoi c'est important :

Sans CORS, un frontend sur https://mon-site.com ne pourrait pas appeler votre API sur https://xxx.ngrok.io

Se configure via un middleware :

app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],  # Autorise toutes les origines
    allow_methods=["GET", "POST"]  # Méthodes autorisées
)



6. Swagger/OpenAPI

À quoi ça sert ?

Standard pour documenter les APIs + interface interactive.

Avantages :

Généré automatiquement par FastAPI

Accessible via /docs ou /redoc

Permet de tester l'API directement depuis le navigateur

```python
from fastapi import FastAPI, HTTPException, Path
from dataclasses import dataclass, asdict
from typing import Union
import json
import math
from fastapi.middleware.cors import CORSMiddleware
import nest_asyncio
from pyngrok import ngrok
import uvicorn
```

```python
# Allow nested events loops
nest_asyncio.apply()
#Uvicorn (le serveur qui exécute FastAPI) utilise asyncio pour gérer les requêtes de manière asynchrone.
#Sans nest_asyncio, Colab bloquerait son démarrage avec l'erreur mentionnée ci-dessus.

app = FastAPI() #instantiate
#FastAPI crée en arrière-plan :
#Une application ASGI (standard Python pour les serveurs web asynchrones)
#Un routeur pour diriger les requêtes vers les bonnes fonctions
#Un système de dépendances pour l'injection de paramètres

# configure CORS
# Configuration CORS
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)


#DataClass book
@dataclass
class Book:
  author: str
  country: str
  image_link : str
  language: str
  link: str
  pages: int
  title: str
  year: int
  id : int


##### API Routes #
# test
@app.get("/test")
def test_route():
    return {"message": "L'API fonctionne !", "status": "OK"}

# get total books
@app.get("/total_books")
def get_total_books() -> dict: #return a dict
  try:
    total_books = {"total number of books": len(data_dict)}
    return total_books
  except Exception as e:
    return {f'error: {str(e)}'}

# get book info by id
@app.get("/book_info/{id}")
def get_book_info(id: int = Path(ge=0)) -> dict: #path for saying it should be >= 0
  if id not in data_dict:
    raise HTTPException(status_code = 404, detail= "This book is not in the list")
  else:
    return data_dict[id]

@app.get("/get_all")
def get_all() -> list[dict]:
  list_books = []
  for id in data_dict:
    list_books.append(data_dict[id])
  return list_books

@app.post("/book")
def add_new_book(book:Book) -> Book:
  if book.id in data_dict:
    raise HTTPException(status_code=404, detail = "This book already in the list")
  data_dict[book.id] = asdict(book)
  return book

@app.get("/advanced_research")
def advanced_research(author: str = None,
                      year:int =None,
                      title:str = None,
                      pages_min: int =0
                      ) -> Union[list[dict],None]:
  filtered_list = filtered_list = list(data_dict.values())

  if author is not None:
    filtered_list = [book for book in filtered_list if book["author"] == author]

  if year is not None:
    filtered_list = [book for book in filtered_list if book["year"] == year]

  if title is not None:
    filtered_list = [book for book in filtered_list if book["title"] == title]

  if pages_min > 0 :
    filtered_list = [book for book in filtered_list if book["pages"] >= pages_min]

  return filtered_list



#####

# Configure Ngrok
NGROK_AUTH_TOKEN = "2vXOrkWUIwGx1xbHr5D3bateLAf_7cpRWRdsP9pzpLxP6mz23" #https://dashboard.ngrok.com/get-started/your-authtoken
ngrok.set_auth_token(NGROK_AUTH_TOKEN)

# Launch tunnel Ngrok
public_url = ngrok.connect(8000).public_url
print(f"Your Books api is accessible at {public_url}/docs")

# lauch fast API server
uvicorn.run(app, host="0.0.0.0", port=8000, log_level="debug")


#works well

```
