```python
# Empty list
texts = []
labels = []
```

```python
# strip()Enlève les espaces et autres caractères de fin de ligne (comme \n,
# \r, etc.) des deux côtés de la chaîne de caractères, c'est-à-dire aussi
# bien au début qu'à la fin.
# rstrip() Enlève les espaces et les caractères de fin de ligne uniquement à
# la fin (à droite) de la chaîne.

for label, f in enumerate(archives):
  print(f"{f} corresponds to {label}")

  with open(f, "r", encoding ='utf-8') as archive:
    for line in archive:
      print(line)
      line = line.rstrip().lower() #drop /n and lower
      line = line.translate(str.maketrans('','', string.punctuation)) # drop punctuation
      texts.append(line)
      labels.append(label)



```

```python
"""
"<unk>" is a convention that we use in NLP to represent unknowns words.
In this case, we assign 0 to this special word
"""

indice = 1
indice_words = {'unk' : 0}
```

```python
for text in train_text:
  print(text)
  tokens = text.split()
  for token in tokens:
    print(token)
    if token not in indice_words:
      indice_words[token] = indice
      indice += 1
```

```python
# convert in int
train_text_int = []
test_text_int = []

for text in train_text: # for each text
  tokens = text.split() # tokenize the whole sentence
  entire_line = [indice_words[token] for token in tokens] #for each words in this sentence, extract its indice from indice_words previously built
  train_text_int.append(entire_line)#append the list of numbers (indices) to finally get a list of numbers equivalent to the sentences (which are list of words)

```

```python
V = len(indice_words)

# Let's built (remembering the theory of markov chains, transition matrix and probability of transistionning from a state to another)
#And remember that we have problems filling with 0 in probas so fill with one
#Benedetti's transition matrix and probas of states
A0 = np.ones((V,V)) #V² size matrix of 1
pi0 = np.ones(V) #vector of V diomensions

#Neruda's transition matrix and probas of states
A1 = np.ones((V,V)) #V² size matrix of 1
pi1 = np.ones(V) #vector of V diomensions
```

```python
 #ex of tokens extracted from a text_as_int
 #[367, 63, 28, 63, 323, 842, 13, 304],
 #[503, 503, 65, 0, 29, 71, 1200],

def compute_counts(text_as_int, A, pi):
  for tokens in text_as_int: # tokens represent the list of ints corresponding to a sentence
    last_indice = None #for first indice of the sequence (367 and 503 for the 3 ex)
    for indice in tokens: #367,63,28...
      #we are in first word of the sequence
      if last_indice is None:
        pi[indice] += 1 #proba of the word indice 367
      else:
        A[last_indice, indice]  += 1 # proba of transition 367->63
      last_indice = indice #next round 63->28
```

```python
# [t for t, y in zip(train_text_int, Y_train) if y == 0] creates a list of text
# sequences (likely represented as lists of integers, where each integer
# corresponds to a word) from train_text_int where the corresponding label in
# Y_train is 0.
# zip(train_text_int, Y_train) pairs each text sequence in train_text_int with
# its corresponding label in Y_train
compute_counts([ t for t, y in zip(train_text_int,Y_train) if y==0 ], A0, pi0)
compute_counts([ t for t, y in zip(train_text_int,Y_train) if y==1 ], A1, pi1)
```

```python
# Now we have numbers corresponding to the number of times the word appears
# (resp. the sequence of words appear) for pi (resp. Ai)
#We then have to normalize to get proper probabilities

A0 /= A0.sum(axis = 1, keepdims=True)
pi0 /= pi0.sum()

A1 /= A1.sum(axis = 1, keepdims=True)
pi1 /= pi1.sum()
```

```python
#now put log to have numbers not 0.0003156 something

logA0 = np.log(A0)
logpi0 =np.log(pi0)
logA1 = np.log(A1)
logpi1 =np.log(pi1)
```

```python
count0 = sum(y == 0 for y in Y_train) # count labels of 1 in Ytrain
count1 = sum(y == 1 for y in Y_train) # labels 0
total = len(Y_train) # count0 +1
p0 = count0 / total # proba a priori of label 0
p1 = count1 / total # same for 1
logp0 = np.log(p0) #log of proba label 0
logp1 = np.log(p1) # same for 1
p0, p1 #75% benedetti 25% neruda
```

```python
# classifier building
class Classifier:
  def __init__(self, logAs, logpis, logpriors):
    self.logAs = logAs
    self.logpis = logpis
    self.logpriors = logpriors
    self.K = len(logpriors) # number of labels (len([0,1]) = 2 in our case)


  def _compute_log_likelihood(self, input_, class_): #proba of a piece of text (input_) to come from a class (class_)
    logA = self.logAs[class_] #Matrix of transition of the class (class_)
    logpi = self.logpis[class_] # same idea for prob

    last_indice = None
    logprob = 0

    for indice in input_:
      if last_indice is None:
        # First token of the sequence
        logprob += logpi[indice]
      else:
        #compute proba of trnsition froml previous word to current one
        logprob += logA[last_indice, indice]

      # Actualize last_indice for next iteration
      last_indice = indice
    return logprob



  def predict(self, inputs):
    predictions = np.zeros(len(inputs))
    for i, input_ in enumerate(inputs):
      #compute the log proba posterior for each class (label)
      posteriors = [self._compute_log_likelihood(input_, c) + self.logpriors[c] for c in range(self.K)] # o ne preidction for each class and we take the higher
      pred = np.argmax(posteriors)
      predictions[i] = pred
    return predictions
```

```python
# each arrangment has to be sorted
clf = Classifier(logAs=[logA0, logA1] ,logpis=[logpi0, logpi1] ,logpriors=[logp0, logp1] )

Ptrain = clf.predict(train_text_int)
```

```python
print(f"Train accuracy: {np.mean(Ptrain == Y_train)}")
```
