Word2Vec est un modèle d'apprentissage automatique développé par des chercheurs de Google, destiné à créer des représentations vectorielles (ou "embeddings") de mots en analysant de vastes corpus de texte. L'objectif de Word2Vec est de convertir chaque mot en un vecteur dans un espace de grande dimension, où les mots aux significations similaires sont proches les uns des autres. Cela permet d'utiliser ces représentations pour diverses tâches de traitement du langage naturel (NLP) comme la recherche de similarité de mots, l'analogie de mots, et d'autres.

```python
words = []

for sentence in sentences:
  tokens = sentence.translate(str.maketrans('','', string.punctuation)).split()
  tokens = [word.lower( ) for word in tokens if word.isalpha()] #lwoercase and get rid of numbers
  print(tokens)
  if tokens: #tokens contient au moins un élément avant de l'ajouter à la liste words.
    words.append(tokens)
```

```python
import os
import PyPDF2
from tqdm import tqdm
```

```python
def extract_text_from_pdf(path_file):
  with open(path_file, "rb") as archive:
    reader = PyPDF2.PdfReader(archive)
    text = ""
    for page in range(len(reader.pages)):
        text += reader.pages[page].extract_text()
  return text
```

```python
path_folder = "/content/drive/Othercomputers/Mon ordinateur portable/Data Science/textos"
all_docs = []

for archive in tqdm(os.listdir(path_folder)):
  if archive.endswith(".pdf"):
    full_path = os.path.join(path_folder, archive)
    print(full_path)
    try:
      doc = extract_text_from_pdf(full_path)
      all_docs.append(doc)
    except Exception as e:
      print(f"Error processing {archive}: {e}")

```

```python
all_docs[2]
```

```python
# append : Ajoute un nouvel élément (ici, une liste) comme un élément unique
# à la fin de la liste cible. Cela signifie qu’en utilisant
# append(sentences), chaque document traité par la boucle for ajoutera une
# sous-liste de phrases à all_sentences, créant une liste de listes.

# extend : Ajoute chaque élément de la liste passée en argument directement
# à la liste cible. En utilisant extend(sentences), vous ajouterez chaque
# phrase de sentences comme un élément individuel dans all_sentences, ce qui
# créera une seule liste plate de toutes les phrases de tous les documents.
all_sentences = []
size = 0

for doc in all_docs:
  size = size + len(doc)
  sentences = doc.split(".")
  all_sentences.extend(sentences) #et pas append
```

```python
size #3M de caracteres
print(len(all_sentences)) #45K phrases
```

```python
# same data process as before
words = []

for sentence in all_sentences:
  tokens = sentence.translate(str.maketrans('','', string.punctuation)).split()
  tokens = [word.lower( ) for word in tokens if word.isalpha()] #lwoercase and get rid of numbers
  #print(tokens)
  if tokens: #tokens contient au moins un élément avant de l'ajouter à la liste words.
    words.append(tokens)
```

```python
len(words)
```

```python
# Model word2vec
model = Word2Vec(sentences = words, vector_size=5000, window=5, min_count=1, workers = 4)
```

```python
close_sentences = model.wv.most_similar("fred", topn = 10) #still not the best
close_sentences
```
