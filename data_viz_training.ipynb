Générer un jeu de données temporelles simulées :
365 jours (de 1 à 365).
Une variable y qui suit une tendance linéaire plus un bruit aléatoire.
Utiliser Seaborn pour afficher un lineplot avec les jours sur l'axe des x et la variable y sur l'axe des y.
Ajouter une tendance de régression pour visualiser la relation entre le temps et la variable y à l'aide de sns.regplot().

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# jours
dt = np.linspace(1, 365, num=365)

# tendance linéaire (y = 2 * X + 5)
m = 2  # pente
b = 5  # ordonnée à l'origine
y_tendance = m * dt + b

bruit = np.random.normal(loc=0, scale=20, size=365)

y = y_tendance + bruit


plt.figure(figsize=(12, 6))
sns.lineplot(x=dt, y=y, label="Données simulées avec bruit")
sns.regplot(x=dt, y=y, scatter=False, color="red", line_kws={"linewidth": 2}, label="Régression")
plt.title("Tendance Linéaire avec Bruit Aléatoire", fontsize=16)
plt.xlabel("Jours", fontsize=12)
plt.ylabel("Valeur de Y", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

📌 Objectif : Comparer plusieurs distributions sur un même histogramme.
💡 Indice : Génère trois distributions normales avec des moyennes différentes et superpose-les dans sns.histplot().

```python
data = np.random.normal(size = (500,3), loc = [10, 15, 20], scale = 5)

data

df = pd.DataFrame(data = data, columns = ["var1", "var2", "var3"])
df

#n melt() est utilisée pour réorganiser un DataFrame en transformant des colonnes en lignes.
df_melted = df.melt(var_name="Variable", value_name="Valeur")
df_melted

plt.figure(figsize = (10,8))
sns.histplot(data = df_melted, x = "Valeur", kde = True, alpha = 0.5, bins = 30, hue = "Variable")
plt.title("Comparaison de trois distributions normales", fontsize=16)
plt.xlabel("Valeur", fontsize=12)
plt.ylabel("Densité", fontsize=12)
plt.grid(True)
```

📌 Objectif : Comparer la distribution de deux groupes (A et B) sur un même graphique.
💡 Indice : Combine sns.boxplot() et sns.violinplot() en superposant les plots avec alpha.
🔹 Concepts : hue, split=True, inner="quart".

```python
data = np.random.uniform(low = 25, high =40 , size = (1000,2))
df = pd.DataFrame(data= data, columns = ["A","B"])
df_melted = pd.melt(df,value_vars= ["A","B"], var_name = 'group')
df_melted

plt.figure(figsize = (10,8))
sns.violinplot(x="group", y="value", data=df_melted, hue="group", split=True, inner="quart", alpha=0.6)
sns.boxplot(x="group", y="value", data=df_melted, width=0.2, showcaps=False, boxprops={"facecolor": "none"}, whiskerprops={'linewidth': 2})


plt.title("Comparaison de la distribution des groupes A et B")
#plt.legend([], [], frameon=False)  # Cacher la légende répétée
plt.show()
```

📌 Objectif : Visualiser des points en 3D.
💡 Indice : Utilise mpl_toolkits.mplot3d et scatter().

```python
x = np.random.normal(size = 100, loc =10, scale = 20)
y = np.random.normal(size = 100, loc =20, scale = 10)
z = np.random.normal(size = 100, loc =20, scale = 20)
#col = np.random.randint(size = 100, low = 0, high = 10)
col = z

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')#active le mode 3D


sc = ax.scatter3D(x, y, z, c=col, cmap='viridis', s=50, edgecolor='k')

ax.set_xlabel("Axe X")
ax.set_ylabel("Axe Y")
ax.set_zlabel("Axe Z")
ax.set_title("Visualisation 3D des points")


ax.view_init(elev=20, azim=60)  # Elevation (haut/bas) et azimuth (rotation)

# 🔹 Ajouter une barre de couleur
fig.colorbar(sc, ax=ax, label="Valeur de Z")

# 🔹 Afficher le graphique
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Données
dico = {'nom': ["Kylian Mbappé", "Antoine Dupont", "LeBron James", "Kevin Durant", "Jonny Wilkinson"],
        'Physique': [4, 6, 10, 5, 7],
        'Vitesse': [10, 6, 7, 6, 5],
        'Mental': [6, 9, 10, 8, 10]
}
df = pd.DataFrame(data=dico)

# Normaliser les données
columns_to_normalize = ['Physique', 'Vitesse', 'Mental']

# Normalisation des données pour chaque athlète
df_normalized = df.copy()
df_normalized[columns_to_normalize] = df[columns_to_normalize].div(df[columns_to_normalize].max()) # on divise chaque colonne par son max, pour normalise

# Préparer le graphique polaire
categories = ['Physique', 'Vitesse', 'Mental']  # catégories de comparaison
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()#  génère une liste d'angles allant de 0 à 2pi
# angles: [0.0, 2.0943951023931953, 4.1887902047863905] dans notre cas

# Créer un graphique polaire
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'projection': 'polar'})
# ax : Cela correspond à un objet Axes, qui est la zone de dessin où les
# données sont tracées. Un "axe" est en fait le système de coordonnées sur
# lequel les courbes, points, etc., sont dessinés. Ici, puisque
# subplot_kw={'projection': 'polar'} est spécifié, l'axe est un axe polaire


# Ajouter les données pour chaque athlète
for athlete in df_normalized['nom']:
    athlete_data = df_normalized[df_normalized['nom'] == athlete][columns_to_normalize].values.flatten().tolist()
    # ici on récupère les athletes data qui sont ses 3 caractéristiques, pour chaque athlete

    ax.plot(angles, athlete_data, label=athlete)  #trace une ligne pour cet athlète sur le graphique polaire, avec son label
    ax.fill(angles, athlete_data, alpha=0.25)  # remplit la zone sous la ligne pour cet athlète avec une couleur translucide (alpha=0.25)


ax.set_xticks(angles)
ax.set_xticklabels(categories)

ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
ax.set_title("Comparaison des profils des athlètes", size=16)

plt.show()
```
