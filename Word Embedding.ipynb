```python
import pandas as pd
import gensim
```

```python
#pre trained model vectors (each row is composed by one word and its vector of 300 components)
vectores = gensim.models.KeyedVectors.load_word2vec_format('/content/drive/Othercomputers/Mon ordinateur portable/Data Science/SBW-vectors-300-min5.txt')
```

```python
def analogas(v1, v2, v3):
  similitud = vectores.most_similar(positive=[v1, v3], negative=[v2])
  print(f"{v1} es a {v2} como {similitud[0][0]} es a {v3}")
```

```python
analogas('países', 'Naciones', 'información')
```

```python
list(vectores.key_to_index.keys())[:100] # pour voir les datas
```

```python
analogas("alemania", "alemán", "frances")
```

```python
analogas("vaca", "leche", "huevo")
```

```python
def cercanos(v):
  vecinos= vectores.most_similar(positive = [v])
  print(f"vecinos de: {v}")
  for word, score in vecinos:#permet de parcourir les paires (mot, score) qui sont retournées par vectores.most_similar(positive=[v])
    print(f"{word} : {score}")
```

```python
cercanos("rey")
```

```python
cercanos("Argentina")
```
