```python
initial_word = {}
first_order = {}
second_order = {}
```

```python
def remove_punctuation(s):
  return s.translate(str.maketrans('','', string.punctuation))


def add2dict(d, k, v): #dict, key, value
  if k not in d:
    d[k] = []
  d[k].append(v)

```

```python
with open(file, "r", encoding = 'utf-8') as archive:
  for line in archive:
    print(line)
    tokens = remove_punctuation(line.rstrip().lower()).split() # remove punct, remove sapce at the end, lower, divide in words (tokens)
    T = len(tokens) #size of the sentece
    print(f"size of the sentence: {T}")
    for i in range(T): # 4 cases: i=0 (first word), i=T-1 last word, if i = 1 (second word), other words between second and last
      token = tokens[i]
      if i == 0 : # first case initialise inital words only
        initial_word[token] = initial_word.get(token, 0) + 1 #get the value of the key token or 0 if doesnt exist
      else:
        t_1 = tokens[i-1] #previous word of the sentence
        if i == T-1: # Last cause we started at indice 0 (0 -> T-1)
          add2dict(second_order, (t_1, token), "END")
        if i == 1:
          add2dict(first_order, t_1, token)
        else:
          t_2 = tokens[i-2]
          add2dict(second_order, (t_2,t_1), token)

```

```python
# First part is now done

#Normalize
tot_initial_word = sum(initial_word.values()) #we directly sum the values of the dict #483
tot_initial_word
for t, c in initial_word.items():
  initial_word[t] = c / tot_initial_word
```

```python
 #same idea for others dicts
 #ex
 #'para' : ['sobrevivirme','que','tus','que','tus','que','mi','tu']
 def list2dict(ts):
  d = {} #empty dict
  n = len(ts) #how many items

  # Occurence of each element in the list
  for t in ts:
    d[t] = d.get(t, 0) + 1

  #convert to relative probabilities
  for t, c in d.items():
    d[t] = c / n
  return d #return dict of probas
```

```python
for t_1, ts in first_order.items():
  first_order[t_1] = list2dict(ts)
first_order
```

```python
for k, ts in second_order.items():
  second_order[k] = list2dict(ts)
```

```python
def word_exemple(d, show):
  #Generate a random number [0,1]
  p0 = np.random.random()
  if (show == 1):
    print(f"p0 = {p0}")

  #Initialize a variable to compute the cumulative probability going accross the words of the list
  cumul = 0
  if (show == 1):
    print(f"cumulative proba: {cumul}")

  # get each key t and its proba p in the dict d
  for t, p in d.items():
    cumul += p #add the proba of current word to the cumul
    if (show == 1):
      print(f"item: {t} of proba: {p}")
      print(f"cumulative proba: {cumul}")

    if p0 < cumul: #if the cumul goes over the random number at the word t, return t
      return t



```

```python
def generador(size):
  for i in range(size):
    sentence = []
    #first word
    word0 = word_exemple(initial_word,0)
    sentence.append(word0)

    #second word
    word1 = word_exemple(first_order[word0], 0)
    sentence.append(word1)

    # second order till the end
    while True:
      word2 = word_exemple(second_order[(word0,word1)], 0)
      if word2 == "END":
        break
      sentence.append(word2)
      word0 = word1
      word1 = word2
    print(' '.join(sentence))
```
