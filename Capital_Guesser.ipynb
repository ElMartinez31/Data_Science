The goal of this project is to build  little game, a capital guesser, where the probability of occurence of a country is based on its population.


```python
import requests
import numpy as np
import pandas as pd

```

```python
# First lest's scrape data

url = "https://restcountries.com/v3.1/all"

try:
  response = requests.get(url)
  countries = response.json()


except requests.exceptions.RequestException as e:
    # Capturer toutes les exceptions liées à la requête (erreurs de connexion, timeout, etc.)
    print(f"Erreur lors de la requête : {e}")

except ValueError:
    # Capturer des erreurs lors du parsing JSON
    print("Erreur lors du parsing des données JSON.")

except KeyError as e:
    # Capturer des erreurs si une clé attendue est manquante dans les données JSON
    print(f"Clé manquante dans les données : {e}")

except Exception as e:
    # Capturer toutes les autres erreurs
    print(f"Une erreur inattendue est survenue : {e}")
```

```python
import unicodedata
def remove_accents(input_str):
    nfkd_form = unicodedata.normalize('NFD', input_str) # Cela décompose les caractères accentués en deux caractères (par exemple, "é" devient "e" et un accent séparé).
    return ''.join([c for c in nfkd_form if unicodedata.category(c) != 'Mn'])
    #Cela filtre les caractères qui sont des marques de diacritiques (accents). Les catégories Mn correspondent aux caractères qui représentent des accents.

remove_accents("éleve")#ok

import random
import math

# Il faut ajouter la population a dico {pays : [capital, population]}
def play_game(dico):
    score = 0

    countries = list(dico.keys())
    capitals = [dico[country][0] for country in countries]
    populations = [dico[country][1] for country in countries]
    smoothed_pop = [math.sqrt(population) for population in populations]

    while True:
        # Select a country based on population as the weight
        selected_country = random.choices(countries, weights=smoothed_pop, k=1)[0]
        print(f"Quelle est la capitale du pays: {selected_country}")

        # Get the answer from the user
        answer = input("Votre réponse: ")

        # Check if the answer is correct
        if remove_accents(answer.lower().strip()) == remove_accents(dico[selected_country][0].lower().strip()):
            print("Bravo, votre score augmente")
            score += 1
        else:
            print("Perdu, vous n'avez plus qu'à tout recommencer")
            print(f"Correct answer was : {dico[selected_country][0]}")
            print(f"Best score is {max(27,score)}, obtained by Martin")
            break

    return score
# Play the game
final_score = play_game(dico)
print(f"Votre score final est: {final_score}")

# Still some improvements to make (get rid of punctuation as for Washington, D.C, and maybe check the similarity between the answer and correct,
# to avoid losing for a letter difference)

```
